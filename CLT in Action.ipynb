```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generating highly skewed population (Exponential Distribution)
np.random.seed(42)
population = np.random.exponential(scale=2, size=10000)

# Taking 1000 samples of size 50 and calculate their names
sample_means = [np.mean(np.random.choice(population, size=500)) for _ in range(5000)]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(population, bins=50, kde=True, color='red')
plt.title("Original Population (Skewed Data)")

plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=50, kde=True, color='blue')
plt.title("Sample Means Distribution (Normal)")

plt.show()
```
